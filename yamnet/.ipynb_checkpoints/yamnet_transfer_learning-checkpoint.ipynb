### Yamnet Predictions auf Rainforest Sample

```python
# The graph is designed for a sampling rate of 16 kHz, but higher rates should work too.
# We also generate scores at a 10 Hz frame rate.
params = yamnet_params.Params(sample_rate=16000, patch_hop_seconds=0.1)
print("Sample rate =", params.sample_rate)
```

```python
# Set up the YAMNet model.
class_names = yamnet_model.class_names('yamnet_class_map.csv')
yamnet = yamnet_model.yamnet_frames_model(params)
yamnet.load_weights('yamnet.h5')
# yamnet.summary()

# yamnet.compile()
# yamnet.fit()
```

```python
import tensorflow_io as tfio
import IPython.display as ipd

# Read in the audio.
wav_file_name = '0a65cc78c.flac'
wav_file_name = '0a65cc78c.wav'
audio = tfio.audio.AudioIOTensor(wav_file_name)
audio = tf.squeeze(audio[:], axis=[-1])
waveform = tfio.audio.resample(audio, 48000, 16000)
waveform = tf.cast(waveform, tf.float32)
waveform = waveform / 32768.0
waveform = waveform[:5*16000]
print(waveform.shape[0] / 16000)
```

```python
# Run the model.
scores, embeddings, spectrogram = yamnet(waveform)
scores = scores.numpy()
spectrogram = spectrogram.numpy()
print(scores.shape, embeddings.shape, spectrogram.shape)
```

```python
# Visualize the results.
plt.figure(figsize=(10, 8))

# Plot the waveform.
plt.subplot(4, 1, 1)
plt.plot(waveform)
plt.xlim([0, len(waveform)])

# Plot the log-mel spectrogram (returned by the model).
plt.subplot(4, 1, 2)
plt.imshow(spectrogram.T, aspect='auto', interpolation='nearest', origin='bottom')

# Plot and label the model output scores for the top-scoring classes.
mean_scores = np.mean(scores, axis=0)
top_N = 10
top_class_indices = np.argsort(mean_scores)[::-1][:top_N]

plt.subplot(4, 1, 3)
plt.imshow(embeddings.numpy().T, aspect='auto', interpolation='nearest')

plt.subplot(4, 1, 4)
plt.imshow(scores[:, top_class_indices].T, aspect='auto', interpolation='nearest', cmap='gray_r')
# Compensate for the patch_window_seconds (0.96s) context window to align with spectrogram.
patch_padding = (params.patch_window_seconds / 2) / params.patch_hop_seconds
plt.xlim([-patch_padding, scores.shape[0] + patch_padding])
# Label the top_N classes.
yticks = range(0, top_N, 1)
plt.yticks(yticks, [class_names[top_class_indices[x]] for x in yticks])
_ = plt.ylim(-0.5 + np.array([top_N, 0]))

#ipd.Audio(waveform, rate=16000)
```

### Transfer learning with Yamnet

```python
# yamnet.summary()
```

```python
# get layers from yamnet
layers = [l for l in yamnet.layers]
input_layer = tf.keras.Input(shape=(96, 64, 1), name='Input')

# attach layer again convolutions on
for i, layer in enumerate(layers[79:]):
    if i == 0:
        x = layer(input_layer)
    else:
        x = layer(x)

# x = tf.keras.layers.Dense(24, activation='sigmoid')(x)

yamnet_tl = tf.keras.Model(inputs=input_layer, outputs=x)
yamnet_tl.summary()
```

```python
# Read in the audio.
wav_file_name = '0a65cc78c.wav'
audio = tfio.audio.AudioIOTensor(wav_file_name)
audio = tf.squeeze(audio[:], axis=[-1])
audio = tfio.audio.resample(audio, 48000, 16000)
audio = tf.cast(audio, tf.float32)
audio = audio / 32768.0

scores, embeddings, spectrogram = yamnet(audio)
print(spectrogram.shape)
```

```python
plt.imshow(spectrogram[:5*96, :].numpy().T, origin='lower')
```

```python
preds = yamnet_tl.predict(np.expand_dims(spectrogram[:96, :], axis=0))
preds.shape
```

```python
import pandas as pd

class_names = yamnet_model.class_names('yamnet_class_map.csv')
df_pred = pd.DataFrame(index=class_names, columns=['pred'])
df_pred['pred'] = preds.T
df_pred = df_pred.sort_values(by='pred', ascending=False)
df_pred
```
